```python
#Create JSON from search term
import subprocess
term = input("Enter search term: ")
filename = term + '.json'
op1 = open(filename, 'w')
subprocess.call(['twarc', 'search', term], stdout=op1)
```

```python
import pandas as pd
tweet_dict = {}
#Build list of tweets from JSON
def build_tweet_list(file_name):
    df = pd.read_json(f"./{file_name}", lines=True)
    text = []
    for i in range(0,len(df)):
        text.append(df['full_text'][i])
    if file_name not in tweet_dict.keys():
        tweet_dict[file_name] = text
    else:
        print('JSON file already in dictionary')
    return tweet_dict.keys()

build_tweet_list('covid19_vaccine.json')
```
